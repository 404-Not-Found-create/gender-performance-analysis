```python
# === Imports ===
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# --- import utils ---
import sys
import os

sys.path.append(os.path.abspath('../utils'))

from utils import (view_ttest, create_pie, create_boxplot)
```

```python
# === DataFrame (StudentsPerformance.csv) ===
students_perfomance = pd.read_csv(r'C:\Users\Admin64\Desktop\WorkSpace\project\Python\gender-performance-analysis\data\StudentsPerformance.csv')
students_perfomance.rename(columns={'parental level of education': 'parental_level_of_education',
                                    'test preparation course': 'test_preparation_course',
                                    'math score': 'math_score',
                                    'writing score': 'writing_score',
                                    'reading score': 'reading_score'}, inplace=True)
students_perfomance.head(6)
```

```python
# === Сhecking the number of women and men in the dataset for a more objective assessment ===
genders = students_perfomance['gender'].value_counts()
create_pie(genders, 'percentage of men and women in the sample')
```

```python
# === Сhecking the educational level of the parents of each of the assessed groups ===
female_parental_level_of_education = students_perfomance[students_perfomance['gender'] == 'female']['parental_level_of_education'].value_counts()
male_parental_level_of_education = students_perfomance[students_perfomance['gender'] == 'male']['parental_level_of_education'].value_counts()
```

```python
create_pie(male_parental_level_of_education, 'male_parental_level_of_education')
```

```python
create_pie(female_parental_level_of_education, 'female_parental_level_of_education')
```

```python
# === Сhecking the level of training ===
female_test_preparation_course = students_perfomance[students_perfomance['gender'] == 'female']['test_preparation_course'].value_counts()
male_test_preparation_course = students_perfomance[students_perfomance['gender'] == 'male']['test_preparation_course'].value_counts()
```

```python
create_pie(male_test_preparation_course, 'male_test_preparation_course')
```

```python
create_pie(female_test_preparation_course, 'female_test_preparation_course')
```

```python
# === Hypothesis testing ===

# --- math score ---
male_math_score = students_perfomance[students_perfomance['gender'] == 'male']['math_score']
female_math_score = students_perfomance[students_perfomance['gender'] == 'female']['math_score']

view_ttest(male_math_score, female_math_score)
create_boxplot(students_perfomance, x='gender', y='math_score', title='The ratio of gender to math scores')
```

```python
# --- reading score ---
male_reading_score = students_perfomance[students_perfomance['gender'] == 'male']['reading_score']
female_reading_score = students_perfomance[students_perfomance['gender'] == 'female']['reading_score']

view_ttest(male_reading_score, female_reading_score)
create_boxplot(students_perfomance, x='gender', y='reading_score', title='The ratio of gender to reading scores')
```

```python
# --- writing score ---
male_writing_score = students_perfomance[students_perfomance['gender'] == 'male']['writing_score']
female_writing_score = students_perfomance[students_perfomance['gender'] == 'female']['writing_score']

view_ttest(male_writing_score, female_writing_score)
create_boxplot(students_perfomance, x='gender', y='writing_score', title='The ratio of gender to writing scores')
```
